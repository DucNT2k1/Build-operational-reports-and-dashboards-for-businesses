Trước tiên, cần xử lí cột 'Level'. Có thể thấy, ở đầu mỗi giá trị trong cột có các ký tự gây khó hiểu như: L8.1,L8.2,... Vì vậy cần loại bỏ các ký tự này.

```python
#Bỏ các ký tự trước dấu '-' của cột Level
sales_1['Level'] = sales_1['Level'].str.split('-').str[-1]
#Xóa kí tự " " ở đầu mỗi giá trị trong cột
sales_1['Level'] = sales_1['Level'].str.strip()
```

Tiếp đến, kiểm tra vùng cột 'Level' khi cột 'Trạng thái' null.

```python
#Xem các giá trị ở cột level ở vùng dữ liệu null
trangthai_null = sales_1[sales_1['Trạng thái'].isna()]
trangthai_null['Level'].value_counts()
```

Các giá trị xuất hiện ở cột 'Level' khi 'Trạng thái' null bao gồm:

*   Khách đã thanh toán toàn bộ    


*   Hủy đơn sau khi chốt    
*   Không nghe máy/Thuê bao

Tiếp theo, kiểm tra khi 'Level' là một trong những giá trị này, 'Trạng thái' sẽ là những giá trị nào.







         
         
                 
                       
          




```python
#FilterLevel
values_to_keep = ['Khách đã thanh toán toàn bộ','Hủy đơn sau khi chốt','Không nghe máy/Thuê bao','Băn khoăn về giá','Trùng','Đã được tư vấn đầy đủ','Không còn nhu cầu']
#FilterData
Filter_data = sales_1[sales_1['Level'].isin(values_to_keep)]
Filter_data
#GroupBy
groupby = Filter_data.groupby(['Level','Trạng thái'])['Trạng thái'].count()
print(groupby)
```

Điền các giá trị null ở cột 'Trạng thái' bằng giá trị xuất hiện nhiều nhất khi quan sát các cặp giá trị của cột 'Level' và 'Trạng thái.

Ví dụ: khi cột 'Level' là "Hủy đơn sau khi chốt", ta sẽ điền vào cột 'Trạng thái' giá trị "Hủy đơn sau khi chốt" do cặp giá trị này xuất hiện nhiều nhất (530 lần). Tương tự với những cột còn lại.

Level:  Băn khoăn về giá --> Suy nghĩ thêm/Trả lời sau

  Hủy đơn sau khi chốt --> Hủy đơn sau khi chốt

  Khách đã thanh toán toàn bộ --> Đặt hàng

  Không còn nhu cầu --> Từ chối

  Không nghe máy/Thuê bao --> Không nghe máy

  Trùng --> Trùng

  Đã được tư vấn đầy đủ --> Suy nghĩ thêm/Trả lời sau



```python
#Điền giá trị null
sales_1['Trạng thái'] = sales_1['Trạng thái'].fillna(sales_1['Trạng thái'].where(~(sales_1['Level'] == 'Băn khoăn về giá'), 'Suy nghĩ thêm/Trả lời sau'))
sales_1['Trạng thái'] = sales_1['Trạng thái'].fillna(sales_1['Trạng thái'].where(~(sales_1['Level'] == 'Hủy đơn sau khi chốt'), 'Hủy đơn sau khi chốt'))
sales_1['Trạng thái'] = sales_1['Trạng thái'].fillna(sales_1['Trạng thái'].where(~(sales_1['Level'] == 'Khách đã thanh toán toàn bộ'), 'Đặt hàng'))
sales_1['Trạng thái'] = sales_1['Trạng thái'].fillna(sales_1['Trạng thái'].where(~(sales_1['Level'] == 'Không còn nhu cầu'), 'Từ chối'))
sales_1['Trạng thái'] = sales_1['Trạng thái'].fillna(sales_1['Trạng thái'].where(~(sales_1['Level'] == 'Không nghe máy/Thuê bao'), 'Không nghe máy'))
sales_1['Trạng thái'] = sales_1['Trạng thái'].fillna(sales_1['Trạng thái'].where(~(sales_1['Level'] == 'Trùng'), 'Trùng'))
sales_1['Trạng thái'] = sales_1['Trạng thái'].fillna(sales_1['Trạng thái'].where(~(sales_1['Level'] == 'Đã được tư vấn đầy đủ'), 'Suy nghĩ thêm/Trả lời sau'))
```
